```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
EXPORT = False
FROM = 0
TO = 0.5
N = 1000
T = (TO-FROM)/N
```

```python
if EXPORT:
    def save(s: str):
        plt.savefig(s, transparent=True, dpi=300)

    # White Text for export
    plt.rcParams.update({
        "ytick.color" : "w",
        "xtick.color" : "w",
        "axes.labelcolor" : "w",
        "axes.edgecolor" : "w",
    })
else:
    def save(s):
        return
```

```python
x = np.linspace(start=FROM, stop=TO, num=N)
y = np.zeros(N)
for n in range(1, 6):
    y += n*np.cos(n*10*2*np.pi*x)

plt.plot(x, y)
save('1.png')
plt.show()
```

```python
xf = np.linspace(start=0.0, stop=1.0/(2.0*T), num=N//2)
yf = np.fft.fft(y)

plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.axis([0, 80, 0, 6])
save('2.png')
plt.show()
```
